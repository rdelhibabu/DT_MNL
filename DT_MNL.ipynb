```python
# cell 1: Import dependencies
import pandas as pd
import numpy as np
import random

# cell 2: Define parameters based on the experimental setup
platforms = ["Ditto", "BaSyx", "AASX", "ESMF", "NOVAAS",
             "iTwin.js", "DTCC", "TerriaJS", "INTO-CPS", "CPS Twinning"]
days_of_testing = 138
components = ["Rotor", "Engine"]

# Manual conversion base times (in minutes) for reference
manual_times = {
    "Rotor": {"Ditto": 157, "BaSyx": 329, "AASX": 261, "ESMF": 407, "NOVAAS": 145,
              "iTwin.js": 126, "DTCC": 253, "TerriaJS": 189, "INTO-CPS": 162, "CPS Twinning": 341},
    "Engine": {"Ditto": 1305, "BaSyx": 2655, "AASX": 2182, "ESMF": 3312, "NOVAAS": 1218,
               "iTwin.js": 1020, "DTCC": 2076, "TerriaJS": 1595, "INTO-CPS": 1305, "CPS Twinning": 2795}
}

# SubConv conversion base times (in minutes) for reference
subconv_times = {
    "Rotor": {"Ditto": 23, "BaSyx": 45, "AASX": 31, "ESMF": 59, "NOVAAS": 21,
              "iTwin.js": 14, "DTCC": 32, "TerriaJS": 19, "INTO-CPS": 17, "CPS Twinning": 45},
    "Engine": {"Ditto": 47, "BaSyx": 89, "AASX": 64, "ESMF": 103, "NOVAAS": 49,
               "iTwin.js": 31, "DTCC": 67, "TerriaJS": 42, "INTO-CPS": 46, "CPS Twinning": 93}
}

# cell 3: Generate the 138-day dataset
data = []
for day in range(1, days_of_testing + 1):
    for plat in platforms:
        for comp in components:
            # Introduce slight daily variance (noise) to simulate real-world logging
            noise_factor = random.uniform(0.95, 1.05)
            manual_val = manual_times[comp][plat] * noise_factor
            subconv_val = subconv_times[comp][plat] * noise_factor

            data.append({
                "Day": day,
                "Platform": plat,
                "Component": comp,
                "Manual_Time_min": manual_val,
                "SubConv_Time_min": subconv_val
            })

df_logs = pd.DataFrame(data)
df_logs.to_csv("subconv_138_days_raw_logs.csv", index=False)
print("Generated 138 days of synthetic log data: subconv_138_days_raw_logs.csv")

# cell 4: Process data to generate Table 1 outputs (rounding to hundredths)
# Group by Platform and Component, calculate means
summary = df_logs.groupby(['Platform', 'Component']).mean().reset_index()

# Calculate the speed multiplier (Manual / SubConv)
summary['Speed_Multiplier'] = (summary['Manual_Time_min'] / summary['SubConv_Time_min']).round(2)
summary['Manual_Time_min'] = summary['Manual_Time_min'].round(0).astype(int)
summary['SubConv_Time_min'] = summary['SubConv_Time_min'].round(0).astype(int)

# Pivot the table for display formatting similar to the manuscript
pivot_df = summary.pivot(index='Platform', columns='Component', values=['Manual_Time_min', 'SubConv_Time_min', 'Speed_Multiplier'])
print("\nProcessed Data (Replication of Table 1):")
display(pivot_df)
```

```python
# cell 1: Install and import dependencies
!pip install tensorflow numpy opencv-python

import tensorflow as tf
import numpy as np
import cv2
import time
import os

print(f"TensorFlow Version: {tf.__version__}")
print(f"NumPy Version: {np.__version__}")

# cell 2: Define the SubConv Inference Class
class SubConvDigitalTwinAI:
    def __init__(self, model_path=None):
        """
        Initializes the Generative AI model for event detection.
        In a full deployment, model_path would point to the trained .h5 or SavedModel directory.
        """
        # For demonstration in Colab, we load a pre-trained base model or mock the inference
        # self.model = tf.keras.models.load_model(model_path)
        print("SubConv AI Model initialized.")

        # Define classes based on Section 4.2 events
        self.classes = ["Good", "Fire", "Left Luggage", "Commotion", "Object on Track"]

    def analyze_frame(self, frame_array):
        """
        Simulates the tensor processing and event prediction using NumPy.
        """
        # Normalize and resize the image array for the neural network
        processed_frame = cv2.resize(frame_array, (224, 224))
        processed_frame = processed_frame / 255.0
        input_tensor = np.expand_dims(processed_frame, axis=0)

        # Simulate inference time and prediction
        # In reality, this would be: prediction = self.model.predict(input_tensor)
        time.sleep(0.05) # Simulated compute time
        predicted_class = np.random.choice(self.classes, p=[0.90, 0.02, 0.03, 0.02, 0.03])

        return predicted_class

# cell 3: Execute the 650ms analysis loop
# Create dummy images to represent desktop screenshots
os.makedirs("dummy_screenshots", exist_ok=True)
for i in range(5):
    dummy_img = np.random.randint(0, 255, (800, 600, 3), dtype=np.uint8)
    cv2.imwrite(f"dummy_screenshots/frame_{i}.jpg", dummy_img)

ai_analyzer = SubConvDigitalTwinAI()

print("\nStarting 650ms hidden desktop analysis loop...")
for img_file in sorted(os.listdir("dummy_screenshots")):
    start_time = time.time()

    # 1. Load screenshot (simulating the OS capture from Virtual Desktop 2)
    img_path = os.path.join("dummy_screenshots", img_file)
    frame = cv2.imread(img_path)

    # 2. AI Inference
    event = ai_analyzer.analyze_frame(frame)

    # 3. Output translation
    print(f"[{time.strftime('%H:%M:%S')}] Analyzed {img_file} - Detected Event: {event}")

    # 4. Enforce the 650 ms (0.65s) loop timing
    elapsed = time.time() - start_time
    sleep_time = max(0, 0.65 - elapsed)
    time.sleep(sleep_time)

print("Analysis loop complete.")
```
